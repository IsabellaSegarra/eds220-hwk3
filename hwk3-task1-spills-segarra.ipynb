```python
# ---Import Libraries---
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

# --- Import Data ---

# Load in NY spills data
fp_spills = os.path.join('data', 'Spill_Incidents_20251106.csv')
spills = pd.read_csv(fp_spills)

# Load in counties data 
# For error while loading with 'os' package on Jupyterlab
os.environ['PROJ_LIB'] = '/opt/anaconda3/share/proj'
fp_counties = os.path.join('data','tl_2023_us_county', 'tl_2023_us_county.shp')
counties = gpd.read_file(fp_counties)

#--- Data Cleaning: Spills---
# Lower case all column names and replace spaces with '_'
spills.columns = (spills.columns
                  .str.lower()
                  .str.replace(' ','_')
                )
print(spills.columns, '\n')

# Update 'spill_date', 'received_date', and 'close_date' columns to be as datetimes
spills[['spill_date', 'received_date', 'close_date']] = spills[
    ['spill_date', 'received_date', 'close_date']].apply(pd.to_datetime)

# -- Data Selecting: Spills --- 
# Select the 'spill_date' column and filter it for select range of dates
petr_23 = spills[ (spills['spill_date'] >= '2023-01-01') & 
                 (spills['spill_date'] <= '2023-10-31') & (spills['material_family'] == 'Petroleum') ] # Select the 'material_family' column, ensuring all materials are petroleum 

#--- Data Aggregation: Spills ---
# Group by 'county' and count the 'spill_number' column. Use the name argument in reset_index to update the column.
spills_per_county = (petr_23.groupby('county')['spill_number'].count().reset_index(name='n_spills'))

# Update 'St Lawrence' to 'St.Lawrence' by locating the with 'iloc'
spills_per_county.loc[spills_per_county['county'] == 'St Lawrence', 'county'] = 'St. Lawrence'

# --- Data Cleaning: Counties ---

# Lower case all column names and replace spaces with '_'
counties.columns = (counties.columns
                  .str.lower()
                  .str.replace(' ','_')
                )
# Refrence the metadata to see state codes and select with the 'statefp' column. 
ny_counties = counties[counties['statefp'] == '36']

# Change 'ny_counties' name column to county 
ny_counties = ny_counties.rename(columns = {'name': 'county'})

# --- Merge Data --- 
ny_counties = pd.merge(ny_counties, spills_per_county, # dataframes merged
        on = 'county', # common column merging on
        how = 'inner') # type of merge 

# Locate the Niagra county column, add a spill to 'n_spills' with +=
ny_counties.loc[ny_counties['county'] == 'Niagara', 'n_spills'] += 1

# --- Map ---  
# Establish empty figure 
fig, ax = plt.subplots(figsize=(8, 6))

# Remove the axis 
ax.axis('off')

# Create choropleth map of number of petroleum spills per county 
ny_counties.plot(
    ax=ax,
    column='n_spills',
    cmap='Reds',
    legend=True,
    edgecolor="0.6",
    linewidth=0.5,
    legend_kwds={
        "shrink": 0.7,
        "label": "Number of petroleum spills",
        "orientation": "horizontal",
        "pad": 0.05
    }
)

ax.set_title('Petroleum Spills in New York Counties', fontsize=15, weight='bold', pad=20)
fig.suptitle('January  October 2023', fontsize=12, style='italic', y=0.92)

plt.show()
```
